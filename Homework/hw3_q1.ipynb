```python
from skimage import data, io, filters, data_dir
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

image = io.imread('seg2.jpg')
plt.imshow(image[:,:])

X = image
X.shape
data = image.reshape(-1,3)
data.shape
```

```python
silhouette_scores = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    silhouette_scores.append(silhouette_score(data, kmeans.labels_))
    
figure = plt.figure(figsize=(15,5))
plt.plot(range(2,20), silhouette_scores)
plt.grid(True)
plt.xticks(ticks = range(2,20))
plt.title('Silhouette_scores vs. k values')
plt.xlabel('k')
plt.ylabel('Silhouette Scores')
```

```python
kmeans3 = KMeans(n_clusters=3).fit(data)
labels = kmeans3.predict(data)
seg_im = labels.reshape(image.shape[0],image.shape[1])

plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(seg_im, cmap = 'gray')
plt.title('Segmented Image for k=3')
plt.subplot(2,2,1)
plt.imshow(seg_im, cmap = 'Reds')
plt.title('Segmented Image for k=3')
plt.show()
```
